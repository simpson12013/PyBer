```python
#Import Dependencies
import csv
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Read csv's and set them to DataFrames
csvpathcity = pd.read_csv('raw_data/city_data.csv')
csvpathride = pd.read_csv('raw_data/ride_data.csv')
city_df = pd.DataFrame(csvpathcity)
ride_df = pd.DataFrame(csvpathride)
```

```python
#Set Universal Color Scheme
colors = ['lightskyblue','lightcoral', 'gold']
```

```python
#Merge DataFrames
city_ride_df = pd.merge(city_df,ride_df,how='outer')
```

```python
# Obtain Series of Cities and Corresponding Type
citytypes_df = city_df.groupby('city')
citytypes = citytypes_df['type'].sum()
```

```python
#Create New City DataFrame Grouped by City
city_grouped = city_df.groupby('city')

#Total Number of Drivers per City
city_drivers= city_grouped['driver_count'].sum()
```

```python
#Create A New City/Ride DataFrame Grouped by City
city_ride_grouped = city_ride_df.groupby('city')

#Find Total Fare per City (Ride_ID and Driver Count are incorrect numbers)
citygroupsum_df = city_ride_grouped.sum()

#Find Average Fare per City (Ride_ID is incorrect)
citygroupavg_df = city_ride_grouped.mean()
averagefarepercity = citygroupavg_df['fare']

#Dollar Formatting for Fare (Striked due to negative interaction with
                            #the Bubble Plot)
#averagefarepercity = averagefarepercity.map('${:,.2f}'.format)
```

```python
#Find Total Fare by Type
city_type_grouped_df = city_ride_df.groupby('type')
citytypesum = city_type_grouped_df.sum()
percentfarebytype = citytypesum['fare']
```

```python
#Find Drivers per Type
typegrouped = city_df.groupby('type')
typedriver_df = typegrouped.sum()
percentdriverbytype = typedriver_df['driver_count']
```

```python
#Find Total Number of Rides Per City
city_riders = city_ride_df['city'].value_counts()
```

```python
#Find Number of Cities in Each Type
type_cities = city_df['type'].value_counts()
```

```python
#Find Number of Rides per Type
type_riders = city_ride_df['type'].value_counts()
```

```python
#Pie Chart for % of Total Fare by City Type

plt.pie(percentfarebytype, labels=percentfarebytype.index, 
        colors = colors, shadow=True, explode = (0.05,0.02,0.05), 
        autopct= '%.2f%%')
plt.title('% of Total Fare by City Type')
plt.show()
```

```python
#Pie Chart for % of Total Rides by City Type

plt.pie(type_riders, labels=type_riders.index, colors = colors, 
        shadow=True, explode = (0.05,0.02,0.05), 
        autopct='%.2f%%')
plt.title('% of Total Rides by City Type')
plt.show()
```

```python
#Pie Chart for % of Total Drivers by City Type

plt.pie(percentdriverbytype, labels=percentdriverbytype.index, 
        colors = colors, shadow=True, explode = (0.05,0.05,0.05), 
        autopct='%.2f%%')
plt.title('% of Total Drivers by City Type')
plt.show()
```

```python
#Create Master DataFrame For the Bubble Plot
data = {'Driver Count': city_drivers, 'Average Fare': averagefarepercity, 
        'Type': citytypes, 'Rides': city_riders }

bubble_df = pd.DataFrame(data=data)

#Split the Master Bubble-DataFrame by Type into Three DataFrames
Urban_bubble_df = bubble_df.loc[bubble_df['Type'] == 'Urban']
Suburban_bubble_df = bubble_df.loc[bubble_df['Type'] == 'Suburban']
Rural_bubble_df = bubble_df.loc[bubble_df['Type'] == 'Rural']
```

```python
#Plot Rural Data into Master Bubble Plot
plt.scatter(x=Rural_bubble_df['Rides'],
            y=Rural_bubble_df['Average Fare'],
            s=Rural_bubble_df['Driver Count']*5, 
            c= colors[2],alpha=0.4,edgecolors='k')

#Plot Suburban Data into Master Bubble Plot
plt.scatter(x=Suburban_bubble_df['Rides'],
            y=Suburban_bubble_df['Average Fare'],
            s=Suburban_bubble_df['Driver Count']*5, 
            c= colors[1],alpha=0.4,edgecolors='k')

#Plot Urban Data into Master Bubble Plot
plt.scatter(x=Urban_bubble_df['Rides'],
            y=Urban_bubble_df['Average Fare'],
            s=Urban_bubble_df['Driver Count']*5, 
            c= colors[0],alpha=0.4,edgecolors='k')

#Format the Bubble Plot
plt.xticks(range(0,40, 5))
plt.yticks(range(15,50, 5))
plt.legend(['Rural', 'Suburban', 'Urban'])
plt.title('PyBer Ride Share Date')
plt.xlabel('Rides Per City')
plt.ylabel('Average Fare Per City')
plt.show()
```
